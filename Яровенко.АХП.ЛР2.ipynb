# **Задание 1**

Заданы три независимые целочисленные неотрицательные случайные величины  $X_1,X_2, X_3$ , необходимо найти математическое ожидание, дисперсию, среднее квадратическое отклонение и коэффициент вариации случайной величины  $X_1+X_2+X_3$  численно и теоретически (из определения и с использованием производящей функции). В таблице ниже по вариантам представлены соответствия значений случайной величины и их вероятности.

```r
Variant<-22
set.seed(Variant)
X1<-sample(c(1:20),5)
X2<-sample(c(3:100),5)
X3<-sample(c(0:40),5)
pp1<-runif(5)
p1<-pp1/sum(pp1)
pp2<-runif(5)
p2<-pp2/sum(pp2)
pp3<-runif(5)
p3<-pp3/sum(pp3)
```

```r
View(data.frame(X1, X2, X3, p1, p2, p3))
```

**Расчеты через формулы**

---



Математическое ожидание:

$$M[X]=\sum_{k=0}^N p_k\cdot x_k$$

где $p_k$ - вероятность, с которой появляется значение $x_k$

```r
math1=sum(X1*p1)
math2=sum(X2*p2)
math3=sum(X3*p3)
mathSum = math1 + math2 + math3
math1; math2; math3; mathSum
```

Дисперсия:

$$D[X]=\sum_{k=0}^N p_k\cdot (x_k-M[X])^2$$

```r
disp1=sum(((X1 - math1)^2)*p1)
disp2=sum(((X2 - math2)^2)*p2)
disp3=sum(((X3 - math3)^2)*p3)
disp = disp1 + disp2 + disp3
disp1;disp2;disp3;disp
```

Среднее квадратичное отклонение:

$$\sigma[X] = \sqrt{D[X]} $$

```r
sd=sqrt(disp)
sd
```

Коэффициент вариации:

$$V[X] = \frac {\sigma[X]} {M[X]} \cdot 100 $$

```r
cov = sd / mathSum * 100
cov
```

Производящая функция:

$$P_{\zeta3}(z) = p_{31} \cdot z^{7} + p_{32} \cdot z^{30} + p_{33} \cdot z^{0} + p_{44} \cdot z^{28} + p_{45} \cdot z^{20}$$

$$P_{\zeta1+\zeta2+\zeta3}(z) = P_{\zeta1}(z) \cdot P_{\zeta2}(z) \cdot P_{\zeta3}(z)$$

```r
p11 <- Data["p1"][1,]
p12 <- Data["p1"][2,]
p13 <- Data["p1"][3,]
p14 <- Data["p1"][4,]
p15 <- Data["p1"][5,]

p21 <- Data["p2"][1,]
p22 <- Data["p2"][2,]
p23 <- Data["p2"][3,]
p24 <- Data["p2"][4,]
p25 <- Data["p2"][5,]

p31 <- Data["p3"][1,]
p32 <- Data["p3"][2,]
p33 <- Data["p3"][3,]
p34 <- Data["p3"][4,]
p35 <- Data["p3"][5,]
```

---



Производящая функция для Пуассона:
$$E_i(t) = e^{\alpha_i(t-1)}$$

# **Задание 5**

Серверная станция состоит из двух модулей питания. Время безотказной работы каждого из них распределено по показательному закону с одинаковыми параметрами:
$$f(t)=\lambda e^{-\lambda t}$$
$t$ - время работы одного модуля до первого отказа.
Второй модуль питания включается сразу же после отказа первого. Определить плотность вероятности времени безотказной работы всей системы, а также математическое ожидание и дисперсию времени безотказной работы.

```r
Variant<-1
alpha<-runif(100)[77]
alpha
```

**По формулам**

---



По условияю время безотказной работы каждого из них распределено по показательному закону с одинаковыми параметрами:
$$f(t)=\lambda e^{-\lambda t}$$
где  $t$ - время работы одного модуля до первого отказа.

Плотность распределения $t$ будет иметь вид:

$$
f_t(t) = \int_{A(t)}^{B(T)}f_{t_1}(t_1)f_{t_2}(t - t_1)dt_1 \\
где A(t) = a + \cfrac{t - (a+d) + |t - (a+d)|}{2}, \\
    B(t) = b + \cfrac{t - (b+c) - |t - (b+c)|}{2}, \\
    a \le t_1 \le b, c \le t_2 \le d
$$

Для данной задачи  a = 0, b = $\infty$, и с = 0, d = $\infty$

$$
A(t) = 0 + \cfrac{t - (0+\infty) + |t - (0+\infty)|}{2} = \cfrac{t - \infty - t + \infty}{2} = 0 \\
B(t) = \infty + \cfrac{t - (\infty+0) - |t - (\infty+0)|}{2} = \infty + \cfrac{t - \infty + t - \infty}{2} = \infty + \cfrac{2t - 2 * \infty}{2} = \infty + t - \infty = t
$$

$$
f_t(t) = \int_{0}^{t}f(t_1)f(t - t_1)dt_1 = \int_{0}^{t}\lambda \cdot e^{-\lambda \cdot t_1} \cdot \lambda \cdot e^{-\lambda \cdot (t-t_1)}dt_1 =
\lambda^2 \cdot \int_{0}^{t} e^{-\lambda \cdot t_1} \cdot e^{-\lambda \cdot (t-t_1)}dt_1 = \\ = \lambda^2 \cdot \int_{0}^{t} e^{-\lambda \cdot t_1 -\lambda \cdot t + \lambda \cdot t_1} dt_1 = \lambda^2 \cdot e^{-\lambda \cdot t} \cdot \int_{0}^{t} dt_1 = \lambda^2 \cdot t \cdot e^{-\lambda \cdot t}
$$

Для расчета математического ожидания и дисперсии воспользуемся преобразованием Лапласа-Стилтьеса:

$$
\beta(s) = \int_{0}^{\infty}e^{-s t} \cdot f(t) dt
$$

Для $ f_t(t): $

$$
\beta_t(s) = \int_{0}^{\infty}e^{-s\cdot t} \cdot f_t(t) dt = \int_{0}^{\infty}e^{-s\cdot t} \cdot \lambda^2 \cdot t \cdot e^{-\lambda \cdot t} dt = \lambda^2 \cdot \int_{0}^{\infty} t \cdot e^{-(\lambda+s) \cdot t} dt = \\ = \lambda^2 \cdot (-\cfrac{t \cdot e^{-(\lambda+s) \cdot t}}{\lambda+s})|_{0}^{\infty} + \cfrac{\lambda^2}{\lambda+s} \cdot \int_{0}^{\infty}e^{-(\lambda+s) \cdot t} dt = 0 - \cfrac{\lambda^2}{(\lambda+s)^2} \cdot e^{-(\lambda+s) \cdot t}|_{0}^{\infty} = \cfrac{\lambda^2}{(\lambda+s)^2}
$$


Математическое ожидаие:

$$
M(t) = - \cfrac{d\beta_{t}(s)}{ds}|_{s=0} = - \cfrac{d(\cfrac{\lambda^2}{(\lambda+s)^2})}{ds}|_{s=0} = -\lambda^2 \cdot \cfrac{d(\cfrac{1}{(\lambda+s)^2})}{ds}|_{s=0} = \cfrac{2\lambda^2}{(\lambda+s)^3}|_{s=0} = \cfrac{2}{\lambda}
$$

```r
M_2x <- 2 * 1/alpha
M_2x
```

Дисперсия:

$$
D(t) = \cfrac{d^2\beta_{\tau}(s)}{ds^2}|_{s=0} - (\cfrac{d\beta_{\tau}(s)}{ds})^2|_{s=0} =
\cfrac{d^2(\cfrac{\lambda^2}{(\lambda+s)^2})}{ds^2}|_{s=0} - \cfrac{4}{\lambda^2} =
-\lambda^2 \cdot \cfrac{d(\cfrac{2}{(\lambda+s)^3})}{ds}|_{s=0} - \cfrac{4}{\lambda^2} = \\ = \lambda^2 \cdot (\cfrac{6}{(\lambda+s)^4})|_{s=0} - \cfrac{4}{\lambda^2} = \cfrac{6}{\lambda^2} - \cfrac{4}{\lambda^2} = \cfrac{2}{\lambda^2}
$$
